Indexing a using Dataframe.ix[ ] :
Early in the development of pandas, there existed another indexer, ix. This indexer was capable of selecting both by label and by integer location. While it was versatile, it caused lots of confusion because itâ€™s not explicit. Sometimes integers can also be labels for rows or columns. Thus there were instances where it was ambiguous. Generally, ix is label based and acts just as the .loc indexer. However, .ix also supports integer type selections (as in .iloc) where passed an integer. This only works where the index of the DataFrame is not integer based .ix will accept any of the inputs of .loc and .iloc.
Note: The .ix indexer has been deprecated in recent versions of Pandas.

### Understanding group by in python

The core syntax can be broken down similar to the Select-From-Groupby-Where clause. Sample code is given below :

Table_name.groupby(['Group'])['Feature'].aggregation()

Table_name to specify the FROM
'Group' is the list of GROUP BY variables
'Feature' is the list of SELECT variables (with or without WHERE condition)
Aggregate() is to specify the aggregation

```python
#Two step query to find oscars by birthlace
oscars_birthplace = oscars.groupby(['race_ethnicity'])
out_birthplace = oscars_birthplace['_unit_id'].count()
print(out_birthplace)
```

```python
### above snippet can be implemented as follows 
oscars_birthplace = oscars.groupby(['birthplace'])['_unit_id'].count()
print (oscars_birthplace)
```

```python
### Adding more levels
#oscars.groupby(['birthplace','_unit_state'])['_unit_id'].count()
# convert to df
df1 = oscars.groupby(['birthplace','_unit_state'])['_unit_id'].count().reset_index()
type(df1)
df1
```

```python
### adding where clause race ethicity is Black
oscars_ethnic= oscars[oscars['race_ethnicity'] == 'Black'].groupby(['birthplace','_unit_state'])['_unit_id','year_of_award_gold'].count().reset_index()
type(oscars_ethnic)
```
